```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
COLUMNS=['CourseName','Provider','Location','DeliveryMethod']
```

```python
dataframes=[]
for page_number in range(1,6):
    response = requests.get(f'https://niccs.us-cert.gov/training/search?keyword=&distance%5Bvalue%5D=&distance%5Bsource_configuration%5D%5Borigin_address%5D=&cae=&items_per_page=20&a=&map-view=1&auto_detect=0&page={page_number}')
    soup = BeautifulSoup(response.content, "html.parser")
    table = soup.find("table") # Find the "table" tag in the page
    rows = table.find_all("tr") # Find all the "tr" tags in the table
    cy_data = []
    for row in rows:
        cells = row.find_all("td") #  Find all the "td" tags in each row
        cells = cells[0:4] # Select the correct columns
        cy_data.append([cell.text for cell in cells]) # For each "td" tag, get the text inside it

    dataframes.append(pd.DataFrame(cy_data, columns=COLUMNS).drop(0,axis=0))

data = pd.concat(dataframes)
```

```python
df=data.to_csv('Training_providers_list.csv',index=False)
```
